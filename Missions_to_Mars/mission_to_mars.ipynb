```python
# Download dependencies
import os
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
import time
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

## Mars Planet Science Scrape

```python
# URL of page to be scraped
url = 'https://redplanetscience.com/'
browser.visit(url)
```

```python
soup = bs(browser.html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
news_title = soup.find_all('div', class_='content_title')[0]
news_title
```

```python
news_p = soup.find_all('div', class_='article_teaser_body')[0]
news_p
```

## JPL Mars Space Images - Featured Image

```python
url = "https://spaceimages-mars.com/"
browser.visit(url)

#scrape page into soup
html = browser.html
soup = bs(html, "html.parser")
```

```python
print(soup.prettify())
```

```python
relative_image_path = soup.find_all('img')[1]['src']
relative_image_path
```

```python
feature_image = url + relative_image_path
feature_image
```

## Galaxy Facts

```python
url = "https://galaxyfacts-mars.com/"
browser.visit(url)

#scrape page into soup
html = browser.html
soup = bs(html, "html.parser")
print(soup.prettify())
```

```python
planet_profile_table = soup.find_all('table', class_= "table table-striped")
planet_profile_table
```

## Mars Hemispheres

```python
url = "https://marshemispheres.com/"
browser.visit(url)
html = browser.html
soup = bs(html, "html.parser")
print(soup.prettify())
```

```python
links = browser.find_by_css('a.product-item img')
print(len(links))

hemisphere_image_urls = []
img_url = {}
images = []

for i,link in enumerate(links): 
    print(i, link)
    browser.find_by_css('a.product-item img')[i].click()
    
    titles = browser.find_by_css('h2.title').text
    print(titles)
    
    html = browser.html
    soup = bs(html, 'html.parser')
    images = url + soup.find('img', class_='wide-image')['src']
   
    img_url['title'] = titles
    img_url['img_url'] = images
    print(img_url)
    hemisphere_image_urls.append(img_url)
    browser.back()
    
print(hemisphere_image_urls)
```
